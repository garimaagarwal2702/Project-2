```python
plt.figure(figsize=(10,5))
sns.boxplot(x="brand",y="price",data=df)
plt.xticks(rotation=90)
plt.title("Brand vs price")
plt.show()
```

```python
counts=df["warranty"].value_counts()
plt.pie(counts,labels=counts.index,autopct='%1.1f%%',data=df)
plt.title("percent of warranty")
plt.show()
```

```python
sorting = df.groupby("brand")["price"].mean().sort_values()
sns.barplot(sorting)
plt.title("Sorted Brands")
plt.xticks(rotation=90)
plt.show()
```

```python
predict = df.groupby("warranty")[[
    "display_size",
    "resolution_width",
    "resolution_height",
    "spec_rating"
]].mean()

predict.plot(kind="bar")
plt.title("laptop estimate")
plt.ylabel("Points")
plt.xticks(rotation=0)
plt.show()
```
